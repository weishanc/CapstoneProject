```python
import pandas as pd
%matplotlib inline
import matplotlib
import seaborn as sns
sns.set()
matplotlib.rcParams['figure.dpi'] = 144
import requests
# import dill
from bs4 import BeautifulSoup
from datetime import datetime

page = requests.get('http://www.disastercenter.com/virginia/crime/index.html') # Use requests.get to download the page.
soup = BeautifulSoup(page.text, "lxml")

dCounty1 = {'County':[],'Year3':[],'Population3':[],'Murder-Man-slaughter':[],
           'Rape':[],'Robbery':[],'AggravatedAssault':[],'TotalViolentCrime':[],
            'Year4':[],'Population4':[],'Burglary':[],
           'LarcenyTheft':[],'VehicleTheft':[],'TotalPropertyCrime':[]}
    
for item in soup.find_all('tr')[0].find_all('a',href=True):

    if 'County' in item.text:
        
        b = item.text.split()
        ind = b.index('County')
        c = b[:ind+1]
        CountyName = " ".join(c)
        
        link_Num = item['href']
        link_str = 'http://www.disastercenter.com/virginia/crime/'+link_Num
        page_county = requests.get(link_str) 
        soup_county = BeautifulSoup(page_county.text, "lxml")
        
        # table 3: Violent Crime Index base upon 100,000 People
        # (Year	 Population	Murder/Man-slaughter	Rape	Robbery	Aggravated Assault	Total Violent Crime)

        
        for i in range(29,39): 
            column_data = soup_county.find_all('tr')[i].find_all('td')

            dCounty1['County'].append(CountyName)
            dCounty1['Year3'].append(column_data[0].text.strip())
            dCounty1['Population3'].append(column_data[1].text.strip())
            dCounty1['Murder-Man-slaughter'].append(column_data[2].text.strip())
            dCounty1['Rape'].append(column_data[3].text.strip())
            dCounty1['Robbery'].append(column_data[4].text.strip())
            dCounty1['AggravatedAssault'].append(column_data[5].text.strip())
            dCounty1['TotalViolentCrime'].append(column_data[6].text.strip())

        # table 4: Property Crime Index per 100,000 People
        # (Year	 Population	Burglary	LarcenyTheft	Vehicle Theft	Total Property Crime)
#         dCounty1 = {'County':[],'Year4':[],'Population':[],'Burglary':[],
#                    'LarcenyTheft':[],'VehicleTheft':[],'TotalPropertyCrime':[]}
        
        for i in range(42,52): 
            column_data = soup_county.find_all('tr')[i].find_all('td')

#             dCounty1['County'].append(CountyName)
            dCounty1['Year4'].append(column_data[0].text.strip())
            dCounty1['Population4'].append(column_data[1].text.strip())
            dCounty1['Burglary'].append(column_data[2].text.strip())
            dCounty1['LarcenyTheft'].append(column_data[3].text.strip())
            dCounty1['VehicleTheft'].append(column_data[4].text.strip())
            dCounty1['TotalPropertyCrime'].append(column_data[5].text.strip())

df = pd.DataFrame(dCounty1)
df.to_csv('VA_CrimeRate.csv')
```
