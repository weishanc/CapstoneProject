```python
dtest = {'A':[1,2,3],'B':[4,5,6]}
dtest['A'].append(data['features'][3]['properties']['north'])
dtest['B'].append(data['features'][3]['properties']['west'])
dftest = pd.DataFrame(dtest)
dftest
```

```python
d = dict({'GEOID':[],'west':[],'east':[],'south':[],'north':[]})
```

```python
for item in data['features']:
    d['GEOID'].append(item['properties']['GEOID'])
    d['west'].append(item['properties']['west'])
    d['east'].append(item['properties']['east'])
    d['south'].append(item['properties']['south'])
    d['north'].append(item['properties']['north'])
    
df = pd.DataFrame(d)
```

```python
df
```

```python
df['lon'] = (df['west']+df['east'])/2
```

```python
df['lat'] = (df['north']+df['south'])/2
```
